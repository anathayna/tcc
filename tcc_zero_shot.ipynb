## <font color="orange">3. Preparação</font>

Esta etapa prepara as entradas (texto e imagem) do banco de dados para que possam ser interpretadas pelo modelo `ViT-L/14`.

São definidos os *prompts* "isso é ódio" e "isso não é ódio" para fazer um primeiro experimento. Esses *prompts* são como rótulos textuais para o teste *zero-shot*, em que o modelo faz a predição se o meme é ou não odioso sem ter sido treinado especificamente para essa tarefa.

Os textos (*prompts*) são *tokenizados*, ou seja, são convertidos em sequências numéricas compatíveis com o modelo.

Esse processo equivale a uma etapa de vetorização, na qual cada token é representado por embeddings aprendidos durante o treinamento do CLIP.

Após a tokenização, os textos são codificados, resultando em vetores normalizados que descrevem semanticamente cada *prompt*.

O mesmo acontece para a imagem, onde é pré-processada e então codificada, produzindo um vetor de características que representa o conteúdo visual no mesmo espaço semântico do texto.

Tanto as representações textuais quanto visuais são normalizadas e comparadas, os resultantes passam por uma função **softmax**, produzindo probabilidades normalizadas para cada classificação: “ódio” e “não ódio”.

Para exemplificar, é selecionada aleatoriamente uma amostra do conjunto de treino:

```python
import torch
import clip
from PIL import Image
import json
import os
import random
import matplotlib.pyplot as plt
from googletrans import Translator

# load the model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-L/14", device=device)

# prepare the inputs
prompts = ["isso é ódio", "isso não é ódio"]
text_inputs = clip.tokenize(prompts).to(device)

# calculate features
with torch.no_grad():
    text_features = model.encode_text(text_inputs)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)

def predict_meme_zero_shot(img_path):
    image = Image.open(img_path).convert("RGB")
    image_input = preprocess(image).unsqueeze(0).to(device)

    with torch.no_grad():
        image_features = model.encode_image(image_input)
        image_features = image_features / image_features.norm(dim=-1, keepdim=True)

        # calculate similarity between image and text features
        similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)

    # the similarity scores are probabilities.
    # similarity[0, 0] corresponds to the probability of "a hateful meme"
    # similarity[0, 1] corresponds to the probability of "a non-hateful meme"
    return similarity.cpu().numpy()[0]

base_path = '/content/model/hateful_memes/'
train_jsonl_path = os.path.join(base_path, "train.jsonl")

with open(train_jsonl_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

sample = json.loads(random.choice(lines))
img_path = os.path.join(base_path, sample["img"])

image = Image.open(img_path).convert("RGB")

translator = Translator()
text_en = sample.get("text", "")
text_pt = translator.translate(text_en, src="en", dest="pt").text if text_en else "(sem texto)"

plt.imshow(image)
plt.axis("off")
plt.figtext(0.5, -0.005, f"Tradução: {text_pt}",
            ha="center", fontsize=10, wrap=True)
plt.show()

probs = predict_meme_zero_shot(img_path)
predicted_label = prompts[probs.argmax()]

prob_str = " | ".join([f"{p}: {prob:.4f}" for p, prob in zip(prompts, probs)])

print(f"Imagem: {sample['img']}")
print(f"Rótulo verdadeiro: {sample.get('label', 'desconhecido')}")
print(f"Probabilidades: [{prob_str}]")
print(f"Predição do modelo: {predicted_label}")
```

## <font color="orange">4. Classificação</font>

Nesta etapa, o CLIP é avaliado em modo zero-shot, sem ajuste fino (fine-tuning), funcionando como baseline para comparação futura.

```python
from tqdm import tqdm

results = []
output_csv = os.path.join(base_path, "resultados_zero_shot.csv")

with open(train_jsonl_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

for line in tqdm(lines, desc="Processando imagens"):
    sample = json.loads(line)
    img_path = os.path.join(base_path, sample["img"])

    try:
        probs = predict_meme_zero_shot(img_path)
        predicted_label = prompts[probs.argmax()]

        results.append({
            "img": sample["img"],
            "text": sample.get("text", ""),
            "true_label": sample.get("label", "desconhecido"),
            "prob_ódio": float(probs[0]),
            "prob_não_ódio": float(probs[1]),
            "predição": predicted_label
        })
    except Exception as e:
        print(f"Erro com {sample['img']}: {e}")

df = pd.DataFrame(results)
df.to_csv(output_csv, index=False, encoding="utf-8")
print(f"\nResultados salvos em: {output_csv}")
```

## <font color="orange">5. Avaliação</font>

Nesta seção, são apresentados os resultados da avaliação do modelo utilizando o CLIP em configuração zero-shot. Para interpretar o desempenho obtido, são empregadas métricas amplamente utilizadas em tarefas de classificação supervisionada, como acurácia, precisão, revocação e F1-score, descritas a seguir.

**Acurácia**:
Mede a proporção total de acertos, ou seja, quantas previsões o modelo acertou em relação ao total de exemplos avaliados.

**Precisão** (Precision):
Indica que entre as amostras que o modelo previu como “ódio”, quantas realmente pertencem a essa classe.

Alta precisão significa poucos falsos positivos (erros ao classificar “não ódio” como “ódio”).

**Revocação** (Recall):
Mede entre os memes que realmente são “ódio”, quantos o modelo conseguiu identificar corretamente.

Alta revocação significa poucos falsos negativos (casos de ódio que o modelo deixou passar).

**F1-score**:
É a média harmônica entre precisão e revocação, equilibrando as duas métricas. É útil quando há diferença de desempenho entre as classes.

**Support**:
Mostra a quantidade de amostras de cada classe utilizadas no cálculo das métricas.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report

df_results = pd.read_csv(os.path.join(base_path, "resultados_zero_shot.csv"))
df_results["true_label"] = df_results["true_label"].astype(int)
df_results["pred_label"] = (df_results["predição"] == "isso é ódio").astype(int)

acc = accuracy_score(df_results["true_label"], df_results["pred_label"])
print(f"Acurácia zero-shot CLIP: {acc:.4f}")

print("\nRelatório de classificação:")
print(classification_report(df_results["true_label"], df_results["pred_label"],
                            target_names=["não ódio", "ódio"]))
```

A matriz de confusão permite visualizar o desempenho do modelo em termos de acertos e erros para cada classe.

Ela mostra quantas amostras de cada classe foram corretamente classificadas e quantas foram confundidas com a outra categoria, facilitando a identificação dos tipos de erro mais comuns.

- **Verdadeiros Positivos (True Positives - TP):** Memes que são de ódio (rótulo verdadeiro = 1) e foram corretamente classificados como ódio (predição = 1).
- **Verdadeiros Negativos (True Negatives - TN):** Memes que não são de ódio (rótulo verdadeiro = 0) e foram corretamente classificados como não ódio (predição = 0).
- **Falsos Positivos (False Positives - FP):** Memes que não são de ódio (rótulo verdadeiro = 0) mas foram classificados incorretamente como ódio (predição = 1).
- **Falsos Negativos (False Negatives - FN):** Memes que são de ódio (rótulo verdadeiro = 1) mas foram classificados incorretamente como não ódio (predição = 0).

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(df_results["true_label"], df_results["pred_label"])
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges",
            xticklabels=["não ódio", "ódio"],
            yticklabels=["não ódio", "ódio"])
plt.xlabel("Predição")
plt.ylabel("Rótulo verdadeiro")
plt.title("Matriz de confusão - CLIP zero-shot")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

tn, fp, fn, tp = confusion_matrix(df_results["true_label"], df_results["pred_label"]).ravel()

total = tn + fp + fn + tp
taxa_acerto = (tp + tn) / total * 100
taxa_erro = (fp + fn) / total * 100

dados_tabela = [
    ["Acertos", "Verdadeiros Positivos (Era Ódio, Predisse Ódio)", tp, f"{tp/total*100:.2f}%"],
    ["Acertos", "Verdadeiros Negativos (Não era Ódio, Predisse Não Ódio)", tn, f"{tn/total*100:.2f}%"],
    ["Erros", "Falsos Positivos (NÃO era Ódio, Predisse Ódio)", fp, f"{fp/total*100:.2f}%"],
    ["Erros", "Falsos Negativos (Era Ódio, Predisse Não Ódio)", fn, f"{fn/total*100:.2f}%"],
    ["Resumo", "Taxa de Acerto Total (Acurácia)", tp + tn, f"{taxa_acerto:.2f}%"],
    ["Resumo", "Taxa de Erro Total", fp + fn, f"{taxa_erro:.2f}%"]
]

df_resumo = pd.DataFrame(dados_tabela, columns=["Categoria", "Descrição", "Quantidade", "Porcentagem"])

def color_rows(row):
    if row['Categoria'] == 'Acertos':
        return ['color: #2ca02c; font-weight: bold'] * len(row)
    elif row['Categoria'] == 'Erros':
        return ['color: #d62728; font-weight: bold'] * len(row)
    else:
        return ['font-weight: bold'] * len(row)

print(f"Total de amostras avaliadas: {total}\n")

styled_df = (
    df_resumo.style
    .hide(axis='index')
    .apply(color_rows, axis=1)
    .set_properties(**{'text-align': 'left'})
    .set_caption("Resumo de Desempenho do Modelo (Zero-Shot)")
)

display(styled_df)
```

A matriz de confusão indica que o modelo ainda tende a classificar uma grande quantidade de exemplos como “ódio”, mesmo quando o rótulo verdadeiro é “não ódio” (3046 Falsos Positivos). Isso indica um viés contínuo para a classe “ódio”.

Por outro lado, ele consegue acertar uma parte significativa dos memes realmente odiosos (1990 Verdadeiros Positivos contra 1029 Falsos Negativos), além de identificar corretamente 2435 casos de "não ódio" (Verdadeiros Negativos). Isso mostra uma razoável capacidade de detecção, embora com baixo equilíbrio geral entre as classes e uma acurácia total (52,06%) fortemente impactada pelo excesso de falsos positivos.

A lógica *fuzzy* permite que, em vez de um rótulo binário (0 ou 1), atribuímos graus de pertinência a diferentes categorias ou conceitos. As próprias probabilidades (`probs`) que o modelo CLIP gera para "isso é ódio" e "isso não é ódio" já podem ser vistas como graus de pertinência a esses conjuntos *fuzzy*.

Por exemplo, se a probabilidade para "isso é ódio" for 0.7, pode-se dizer que o meme tem um grau de 0.7 de pertinência ao conjunto "ódio". A lógica fuzzy permite ir além, definindo **termos linguísticos** e **funções de pertinência** para categorizar a "intensidade" do ódio.


```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats

class UnifiedFuzzyClassifier:
    """
    - Funções de pertinência triangulares com 5 categorias
    - Cálculo de entropia (incerteza do CLIP)
    - Inferência com regras de contexto
    - Defuzzificação
    """
    def __init__(self, none_peak=0.0, low_peak=0.25, medium_peak=0.5, high_peak=0.75, extreme_peak=1.0):
        self.none_peak = none_peak
        self.low_peak = low_peak
        self.medium_peak = medium_peak
        self.high_peak = high_peak
        self.extreme_peak = extreme_peak

    def _triangular(self, x, a, b, c):
        if x <= a or x >= c: return 0.0
        elif a < x <= b: return (x - a) / (b - a) if b != a else 1.0
        else: return (c - x) / (c - b) if c != b else 1.0

    def _fuzzify(self, prob_odio, entropia):
        none = self._triangular(prob_odio, -0.1, self.none_peak, 0.25)
        low = self._triangular(prob_odio, 0.0, self.low_peak, 0.5)
        medium = self._triangular(prob_odio, 0.25, self.medium_peak, 0.75)
        high = self._triangular(prob_odio, 0.5, self.high_peak, 1.0)
        extreme = self._triangular(prob_odio, 0.75, self.extreme_peak, 1.1)

        if entropia > 0.95:
            medium = min(1.0, medium + 0.3)
            none *= 0.7
            low *= 0.7
            high *= 0.7
            extreme *= 0.7

        return {
            'Nenhum Ódio': none,
            'Pouco Ódio': low,
            'Ambiguidade/Moderado': medium,
            'Ódio': high,
            'Alto Grau de Ódio': extreme
        }

    def classify(self, prob_odio, prob_nao_odio):
        entropia = scipy.stats.entropy([prob_odio, prob_nao_odio], base=2)

        membership = self._fuzzify(prob_odio, entropia)

        numerador = (
            self.none_peak * membership['Nenhum Ódio'] +
            self.low_peak * membership['Pouco Ódio'] +
            self.medium_peak * membership['Ambiguidade/Moderado'] +
            self.high_peak * membership['Ódio'] +
            self.extreme_peak * membership['Alto Grau de Ódio']
        )
        denominador = sum(membership.values())
        score_final = numerador / denominador if denominador != 0 else self.medium_peak

        categoria_final = max(membership, key=membership.get)

        return pd.Series({
            'entropia_modelo': entropia,
            'score_fuzzy': score_final,
            'categoria_fuzzy': categoria_final,
            'confianca_fuzzy': membership[categoria_final]
        })

unified_fuzzy_classifier = UnifiedFuzzyClassifier()
```

#### 1. Fuzzificação

É o processo de conversão das probabilidades brutas geradas pelo modelo CLIP (valores *crisp*) em **graus de pertinência**. Em vez de a probabilidade pertencer exclusivamente a uma classe, ela é mapeada em três funções de pertinência triangulares: *Pouco/Nenhum Ódio*, *Ambiguidade/Moderado* e *Alto Grau de Ódio*. Isso permite que uma mesma imagem possua, por exemplo, 30% de pertinência em "Alto Grau" e 70% em "Ambiguidade".

#### 2. Entropia e Regras de Inferência (Mecanismo de Contexto)

A entropia de Shannon quantifica o grau de incerteza nas predições do modelo CLIP. Em uma classificação binária, ela varia de 0 (certeza absoluta) a 1 (dúvida máxima). No contexto de memes multimodais, uma entropia alta (ex: superior a 0.90) geralmente indica a presença de **tensão semântica ou confundidores**, situações em que o texto e a imagem transmitem mensagens contraditórias.

As **Regras de Inferência** utilizam essa entropia para ajustar os graus de pertinência. Por exemplo: *SE a entropia for muito alta (incerteza do CLIP), ENTÃO o grau de pertinência da categoria 'Ambiguidade/Moderado' é artificialmente elevado e os extremos são penalizados*. Essa regra protege o sistema de tomar decisões precipitadas (Falsos Positivos) quando o modelo está confuso.

#### 3. Defuzzificação (*Score Fuzzy*)

É a etapa final onde os graus de pertinência ponderados pelas regras são convertidos de volta em um valor numérico único consolidado, utilizando o método do **Centroide** (Centro de Gravidade).

Diferente da probabilidade bruta do CLIP, que possui um limiar rígido (onde 0.49 e 0.51 geram classificações opostas), o *score fuzzy* atua como um termômetro contínuo. Se o modelo apresenta alta entropia, o sistema "puxa" a pontuação final para o centro, reduzindo drasticamente os Falsos Positivos causados por pequenas margens numéricas.

```python
def plot_funcoes_pertinencia(classifier):
    scores = np.linspace(0, 1, 1000)

    none_vals = [classifier._triangular(s, -0.1, classifier.none_peak, 0.25) for s in scores]
    low_vals = [classifier._triangular(s, 0.0, classifier.low_peak, 0.5) for s in scores]
    medium_vals = [classifier._triangular(s, 0.25, classifier.medium_peak, 0.75) for s in scores]
    high_vals = [classifier._triangular(s, 0.5, classifier.high_peak, 1.0) for s in scores]
    extreme_vals = [classifier._triangular(s, 0.75, classifier.extreme_peak, 1.1) for s in scores]

    plt.figure(figsize=(10, 5))
    plt.plot(scores, none_vals, label='Nenhum Ódio', linewidth=2, color='#1f77b4') # Azul
    plt.plot(scores, low_vals, label='Pouco Ódio', linewidth=2, color='#2ca02c') # Verde
    plt.plot(scores, medium_vals, label='Ambiguidade/Moderado', linewidth=2, color='#ff7f0e') # Laranja
    plt.plot(scores, high_vals, label='Ódio', linewidth=2, color='#d62728') # Vermelho
    plt.plot(scores, extreme_vals, label='Alto Grau de Ódio', linewidth=2, color='#9467bd') # Roxo

    plt.xlabel('Probabilidade de Ódio (Saída do CLIP)', fontsize=12)
    plt.ylabel('Grau de Pertinência', fontsize=12)
    plt.title('Funções de Pertinência - Classificação Fuzzy', fontsize=14)
    plt.legend(loc='upper right')
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

plot_funcoes_pertinencia(unified_fuzzy_classifier)
```

três categorias (funções triangulares)

🟢 Pouco/Nenhum Ódio: 0.25 [0.0 → 0.5] 🟠 Ambiguidade/Moderado: 0.5 [0.2 → 0.75] 🔴 Alto Grau de Ódio: 0.75

A distribuição do dataset (40% ódio multimodal, 10% ódio unimodal, 20% confundidores de texto, 20% de imagem e 10% não odiosos) exige uma análise altamente contextual. O CLIP zero-shot tende a falhar nos confundidores, ancorando-se frequentemente no sinal visual ou em palavras isoladas, sendo enganado pela contradição entre modalidades.

Esse comportamento reforça a importância de abordagens complementares e a necessidade de experimentos futuros, como:

*   **Ajuste Fino (Fine-tuning):** Treinar o modelo CLIP em uma porção do dataset Hateful Memes para adaptar suas representações ao domínio específico de memes odiosos.

*   **Exploração de Prompts:** Experimentar diferentes prompts e formatos de prompt para a classificação zero-shot, buscando formulações que melhor capturem a distinção entre ódio e não ódio no contexto dos memes.

*   **Análise de Erros:** Investigar mais a fundo os casos de falsos positivos e falsos negativos para entender quais características visuais ou textuais estão levando o modelo a cometer esses erros na presença de ironia ou sarcasmo.

*   **Aprimoramento do Sistema Fuzzy:** Refinar as regras de inferência e as funções de pertinência com base no conhecimento extraído da análise de erros.

*   **Bootstrapping:** Aplicar técnicas de reamostragem para gerar múltiplos subconjuntos do dataset, permitindo estimar a incerteza do modelo, identificar características mais relevantes e reduzir o risco de *overfitting*, contribuindo para modelos mais robustos e precisos.

Em resumo, este experimento zero-shot, complementado pelo pipeline fuzzy, serve como uma linha de base.
